```python
# For every customer, display the full name in CAPS and also the length of the customers full name

pd.read_sql_query("""
SELECT 
    UPPER(CONCAT(customer_first_name, ' ', customer_last_name)) AS `name_in_caps`,
    LENGTH(CONCAT(customer_first_name, customer_last_name)) AS `full_name_length`
FROM customer
""",engine)
```
