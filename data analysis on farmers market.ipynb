```python
# Who did better in the exam Roll No 1048 or Roll No 1313 ?

pd.read_sql_query("""
SELECT subject, SUM(marks), COUNT(marks) AS `# of test takers`, MIN(marks), MAX(marks), AVG(marks)
FROM subject
GROUP BY subject
""",rank_engine)
```

```python
# Who did better in the exam Roll No 1048 or Roll No 1313 ?

pd.read_sql_query("""
SELECT *
FROM subject
WHERE roll_no IN (1313, 1048)
""",rank_engine)
```

```python
pd.read_sql_query("""
SELECT *,ROUND(SUM(marks) OVER(ORDER BY marks ASC ROWS BETWEEN CURRENT ROW AND UNBOUNDED FOLLOWING),0) AS `cumulative_sum` FROM subject;
""", rank_engine)
```

```python
pd.read_sql_query("""
SELECT *,SUM(marks) OVER(ORDER BY marks ASC ROWS BETWEEN 1 PRECEDING AND 1 PRECEDING) AS `one above` FROM subject;
""", rank_engine)
```

```python
pd.read_sql_query("""
SELECT transaction_time,customer_id, quantity, SUM(quantity) OVER(ORDER BY transaction_time ASC ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) AS
`cum_sum_of_quantity` FROM customer_purchases WHERE market_date = "2020-08-05" ORDER BY market_date ASC, transaction_time ASC""",engine)
```

```python
pd.read_sql_query("""
SELECT 
    transaction_time,
    customer_id, 
    quantity, 
    ROUND(SUM(quantity) OVER(ORDER BY transaction_time ASC ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW),2) AS `cum_sum_of_quantity`, 
    ROUND(quantity - SUM(quantity) OVER(ORDER BY market_date ASC, transaction_time ASC ROWS BETWEEN 1 PRECEDING AND 1 PRECEDING),2) AS `change_of_quantity_over_transaction` 
FROM customer_purchases 
WHERE market_date = "2020-08-05" 
ORDER BY market_date ASC, transaction_time ASC
""", engine)
```

```python
pd.read_sql_query("""
SELECT *, ROUND(AVG(marks) OVER(PARTITION BY subject ORDER BY marks ASC ROWS BETWEEN 2 PRECEDING AND CURRENT ROW),2) AS `avg_of_last_3_rows` FROM subject 
""",rank_engine)
```
